```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
# create databas use pandas
data = {

        "Product": ["Laptop", "Phone", "Tablet", "Monitor", "Keyboard"],
        "City": ["Riyadh", "Jeddah", "Dammam", "Mecca", "Khobar"],
        "Quantity": [50, 30, 20, 15, 40]


}
df = pd.DataFrame(data)
```

```python
df
```

```python
df.info()
```

```python
# pivoting data
pivot_table = pd.pivot_table(df,
                             values ='Quantity',
                             index = 'City',
                             columns='Product',
                             aggfunc='mean',
                             fill_value=0)
```

```python
pivot_table
```

```python
pivot_table.info()
```

```python
# data groupping
grouping_data = df.groupby('City')['Quantity'].sum().sort_values(ascending=False)
```

```python
grouping_data
```

```python
# aggregation data
aggregation_data = df.groupby('City')['Quantity'].agg(['sum','mean','max','min'])
```

```python
aggregation_data
```
